```python
class Funcionario:

  num_de_func = 0
  qntaumento = 1.05

  def __init__(self, primeiro, ultimo, salar):
    self.primeiro = primeiro
    self.ultimo = ultimo
    self.salar = salar
    self.email = primeiro + ultimo + '@company.com'

#Contando quantos funcionários foram cadastrados    
    Funcionario.num_de_func += 1

  def fullName(self):
    return f'{self.primeiro} {self.ultimo}'


  def aumentoSal(self):
    self.salar = int(self.salar * self.qntaumento) #Ou Funcionario.qntaumento
    return self.salar

#Aqui não tinha nenhum funcionário
print(f'Temos {Funcionario.num_de_func} funcionários na empresa')
func1 = Funcionario('José', 'Armando', 50000)
func2 = Funcionario('Carlos', 'Magno', 33300)
print('Agora', end=' ')
#Agora tem 2 funcionários
print(f'temos {Funcionario.num_de_func} funcionários na empresa')

print('-'*30)

print(f'O primeiro funcionário se chama: {func1.fullName()}', end=' ')
print(f' e o segundo atende pelo nome de {Funcionario.fullName(func2)}')
print(func1.salar)
#Podemos chamar o funcionário e seu salário de duas formas:
print(f'Com o aumento salarial o {func1.fullName()} passa a receber R${func1.aumentoSal()}')
print(f'E o {Funcionario.fullName(func2)} passa a receber R${Funcionario.aumentoSal(func2)}')

print('-'*30)

#Aqui só muda o aumento do func1 se fosse Funcionario mudaria de todos
func1.qntaumento = 1.08

print(Funcionario.qntaumento)
print(func1.qntaumento)
print(func2.qntaumento)
```

```python
class Dev(Funcionario):
#Subscrevendo novo aumento
  qntaumento = 1.10  

  def __init__(self, primeiro, ultimo, salar, prog_lang):               
    Funcionario.__init__(self, primeiro, ultimo, salar)
    self.prog_lang = prog_lang
    
dev1 = Dev('Julia', 'Souza', 2000, 'Java')
dev2 = Dev('Amarildo', 'Costa', 40000, 'C++')

#print(help(Desenvolvedor))

print(f'O e-mail da Dev {dev1.fullName()} é {dev1.email}')
print(f'A linguagem de programação principal do {dev2.fullName()} é {dev2.prog_lang}')

print(f'O salário de Julia é: {dev1.salar}', end=' ')
dev1.aumentoSal()
print(f'Com o aumento passa a ser: {dev1.salar}')

print('-' *60)

#Criando outra classe com herança

class Gerente(Funcionario):
  qntaumento = 1.20
  def __init__(self, primeiro, ultimo, salar, setor, subordinados= None):
    Funcionario.__init__(self, primeiro, ultimo, salar)
    self.setor = setor
    if subordinados is None:
      self.subordinados = []
    else:
      self.subordinados = subordinados

#método para adicionar subordinados
  def add_sub(self, sub):
    if sub not in self.subordinados:
      self.subordinados.append(sub)

#método para deletar subordinados
  def del_sub(self, sub):
    if sub in self.subordinados:
      self.subordinados.remove(sub)

#método para imprimir os funcionários subordinados
  def print_sub(self):
    for sub in self.subordinados:
      print('--->', sub.fullName())


ger1 = Gerente('Carla', 'Joaquina', 7000, 'Comercial', [dev1])
ger2 = Gerente('Mariana', 'Conrado', 8000, 'Cobrança')

print(f'A gerente {ger2.fullName()} é do setor da {ger2.setor}')
ger1.add_sub(dev2)
print('E seus subordinados são:')
ger1.print_sub()


print(f'O salário de Carla é: {ger1.salar}', end=' ')
ger1.aumentoSal()
print(f'e após o aumento será: {ger1.salar}')

print('-'*60)
#Testes lógicos
print(isinstance(ger1, Gerente))
print(issubclass(Dev, Funcionario))
print(issubclass(Dev, Gerente))
```
